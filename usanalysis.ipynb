## 🎯 Objective

The goal of this project is to explore the US trending YouTube videos dataset, clean the data, perform visual analysis, and derive key insights about video performance and audience engagement.


```python
#importing all required libraries
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import os
```

```python
# checking whether the file exists or not and is indeed a file 
#showing the first 5 rows of the dataset
file_path=r"D:\Jupyter Notebook\Youtube_Analysis\data\USvideos.csv" #pathway to clarity 
if not os.path.exists(file_path):
    raise FileNotFoundError(f"The file'{file_path}' was not there.")
print(os.getcwd())
```

```python
df= pd.read_csv(file_path) 
df.head(5)
```

## 📥 Dataset Overview

This dataset contains details about trending YouTube videos in the US, including:
- Video title
- Channel name
- Views, likes, dislikes, comment count
- Video category and publish time

Let’s check its structure and shape.


```python
#showing the shape ,size and information

df.info()
df.shape
df.size
df.isnull().sum()
df.isnull().any()
```

```python
# sorting the values in ascending oredr from A to Z
df.sort_values(by="channel_title",ascending=True)
```

```python
#cleaning the columns
df=df.dropna()
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')\
.str.replace("'", "",regex=False).str.replace('"',"",regex=False) #dataframe asa df
print("cleaned column nmaes:", df.columns.tolist())
```

## 🧹 Data Cleaning

- Removed missing values  
- Cleaned column names (lowercase, underscores)  
- Ready for analysis


```python
#extracting the time based functions from the 'published_time' column
#coverting publis_time into a proper date time format and creates new columns
#we analyzed the time trends by using uploade_year,"uploadeD_month,uploaded hour,uploade day
#i also gave seperate patterrn for the day of the week for weekdays
# for weekend i used a boolean flag to identify wheter the videos was uploaded on a weekend (saturday,sunday)
df['publish_time']=pd.to_datetime(df['publish_time'],errors="coerce")
df['upload_year']=df['publish_time'].dt.year
df['upload_month']=df['publish_time'].dt.month
df['upload_day']=df['publish_time'].dt.day
df['upload_hour']=df['publish_time'].dt.hour
df['weekdays']=df['publish_time'].dt.day_name()
df['is_weekend']=df['weekdays'].isin(['Saturday','Sunday'])
print(df[["weekdays","is_weekend"]].drop_duplicates().sort_values(by="weekdays"))
#it shows the important columsn to verify uploaded time features
df[['title','publish_time','upload_year','upload_month','upload_hour','weekdays','is_weekend']]
```

```python
# giving statistics for numeric columns like
# count, mean, std, min, max, and quartile values like 25%,50%,75%)
df.describe()
```

```python
#counting how manyy videos appears in each category
plt.figure(figsize=(8,4))
sns.countplot(y="category_id",data=df,order=df["category_id"].value_counts().index)
plt.title("Distribution of Trending videos by category id")
plt.xlabel("Number of Videos")
plt.ylabel("category ID")
plt.show()
```

```python
#finding which channels had the most trending videos
top_channels =df["channel_title"].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=top_channels.values,y=top_channels.index)
plt.title("Top 10 Trending youtube channels")
plt.xlabel("number of Trnding videos")
plt.ylabel("channel title")
plt.show()
```

```python
#videos uploaded per year
# There are no videos uploaded from 2006 to 2016
videos_per_year = df['upload_year'].value_counts().sort_index()
videos_per_year = videos_per_year[videos_per_year >0]
colors=sns.color_palette("Paired",n_colors=len(videos_per_year))
plt.figure(figsize=(8,4))
sns.barplot(x=videos_per_year.index,y=videos_per_year.values,palette=colors)
plt.title("Trending Videos Skyrocketed in 2018")
plt.xlabel("year")
plt.ylabel("no.of videos")
plt.grid(False)
plt.xticks(rotation=5)
plt.show()
```

```python
#Using scatterplot to combined both top & bottom 10 by views and likes 
top_10_views=df.sort_values(by="views",ascending=False).head(10)
top_10_views["group"]="Top 10"
bottom_10_views=df.sort_values(by="views",ascending=True).head(10)
bottom_10_views["group"] ="Bottom 10"
combined_df =pd.concat([top_10_views,bottom_10_views])
sns.scatterplot(x="views",y="likes",
                hue="group",data=combined_df,style="group",
                s=80,palette={"Top 10": "Green", "Bottom 10":"pink"})
plt.title("Views vs Likes for Top and Bottom 10 videos",fontsize=12)
plt.xlabel("Views",fontsize=6)
plt.ylabel("Likes",fontsize=6)
plt.tight_layout()
plt.show()
    
```

```python
#code to create the bottom 10 videos Dataframe
bottom_10_views=df.sort_values(by="views",ascending=True).head(10)
bottom_10_views["group"]="Bottom 10"
plt.figure(figsize=(10,4))
sns.barplot(x="views",y="likes",data=bottom_10_views)
plt.title("Views and Likes of the Bottom 10 Trending Videos")
plt.xlabel("views")
plt.ylabel("Likes")
plt.show()
```

```python

#sorting top10 most commented videos in descending order of comment count and also dropping duplicate titles 
#using a bar chart for the most commented videos
top_commented = df.sort_values(by='comment_count',ascending=False).drop_duplicates(subset='title').head(10)
plt.figure(figsize=(10,4))
sns.barplot(x='comment_count',y='title',data=top_commented)
plt.title("Top 10 most commented videos",fontsize=12)
plt.xlabel("Total comments")
plt.ylabel("video title")
plt.show()
```

```python
total_likes=df['likes'].sum()
total_dislikes=df['dislikes'].sum()
total_comments=df['comment_count'].sum()
labels=['likes', 'dislikes', 'comment']
values=[total_likes,total_dislikes,total_comments]
colors=['#ffd700','#99cc33','#66b3ff']
plt.figure(figsize=(4,2))
plt.barh(labels,values,color=colors)
plt.xlabel("Total Count")
plt.ylabel("Metric")
plt.show()
```

```python
#counting weekend vs weekdays uploads with clear labelling 
upload_counts = df["is_weekend"].value_counts().rename({True:"Weekend",False:"weekdays"})
plt.figure(figsize=(4,4))
plt.pie(upload_counts.values,labels=upload_counts.index,colors=["skyblue","violet"], autopct='%1.1f%%')
plt.title("Weekday Vs Weekend Uploads", fontsize=12)
plt.show()
```

```python
#checking the correlation between views,likes,dilikes,and commentcounts
plt.figure(figsize=(10,5))
sns.heatmap(df[['views', 'likes', 'dislikes', 'comment_count']].corr(), annot=True,cmap=("winter"))
plt.title("Correlation between views,likes,dislikes and comments")
plt.show()
```
